# Pegasus WMS Workflow Example

This notebook demonstrates a complete **Pegasus WMS** workflow for a data analysis pipeline.

## Pipeline Overview

```
              ┌─────────────────┐
              │  input_data.csv │
              └────────┬────────┘
                       │
              ┌────────▼────────┐
              │   Preprocess    │
              └────────┬────────┘
                       │
          ┌────────────┼────────────┐
          ▼            ▼            ▼
   ┌──────────┐ ┌──────────┐ ┌──────────┐
   │  Model A │ │  Model B │ │  Model C │
   └────┬─────┘ └────┬─────┘ └────┬─────┘
          └────────────┼────────────┘
                       ▼
              ┌─────────────────┐
              │   Evaluate      │  ← Fan-in merge
              └────────┬────────┘
                       │
              ┌────────▼────────┐
              │  final_report   │ ← Staged out
              └─────────────────┘
```

### Key Pegasus Concepts Demonstrated:
- **File objects** shared between jobs for DAG linking
- **Automatic dependency inference** via `infer_dependencies=True`
- **Parallel execution** (3 models trained in parallel)
- **Fan-in merge** (evaluation collects all model outputs)
- **Site/Transformation/Replica catalogs**
- **Container support** (Docker)
- **Stage-out** for final outputs

````python
# ────────────────────────────────────────────────────────────
# Cell 1: Imports and setup
# ────────────────────────────────────────────────────────────
import os
from Pegasus.api import *

# Use a persistent work directory
WORK_DIR = "/home/pegasus/work/example-workflow"
os.makedirs(WORK_DIR, exist_ok=True)
os.chdir(WORK_DIR)

print(f"Working directory: {WORK_DIR}")
````

````python
# ────────────────────────────────────────────────────────────
# Cell 2: Define the Docker container
# ────────────────────────────────────────────────────────────
container = Container(
    "python-ml",
    Container.DOCKER,
    image="python:3.11-slim",
    mounts=["{WORK_DIR}:/opt/work:rw".format(WORK_DIR=WORK_DIR)]
)

print(f"Container defined: {container.name} -> {container.image}")
````

````python
# ────────────────────────────────────────────────────────────
# Cell 3: Site Catalog — defines execution & storage locations
# ────────────────────────────────────────────────────────────
sc = SiteCatalog()

local_site = Site("local", arch=Arch.X86_64, os_type=OS.LINUX)
local_site.add_directories(
    Directory(Directory.SHARED_STORAGE, path=WORK_DIR)
        .add_file(File("input_data.csv"))
)

sc.add_sites(local_site)
sc.write()
print("[SC] Site catalog written → sites.yml")
print(sc)
````

````python
# ────────────────────────────────────────────────────────────
# Cell 4: Transformation Catalog — registers executables
# ────────────────────────────────────────────────────────────
tc = TransformationCatalog()

# Each transformation maps to an executable script
# (is_stageable=False means it lives in the container)
preprocess_trans = Transformation(
    "preprocess",
    site="local",
    pfn="/opt/work/bin/preprocess.py",
    is_stageable=False,
    container=container
)

train_trans = Transformation(
    "train",
    site="local",
    pfn="/opt/work/bin/train.py",
    is_stageable=False,
    container=container
)

evaluate_trans = Transformation(
    "evaluate",
    site="local",
    pfn="/opt/work/bin/evaluate.py",
    is_stageable=False,
    container=container
)

tc.add_transformations(preprocess_trans, train_trans, evaluate_trans)
tc.write()
print("[TC] Transformation catalog written → tc.yml")
print(f"      Registered: {[t.name for t in [preprocess_trans, train_trans, evaluate_trans]]}")
````

````python
# ────────────────────────────────────────────────────────────
# Cell 5: Replica Catalog — maps logical files to physical paths
# ────────────────────────────────────────────────────────────
rc = ReplicaCatalog()

input_file = File("input_data.csv")
rc.add_replica("local", "input_data.csv", os.path.join(WORK_DIR, "input_data.csv"))

rc.write()
print("[RC] Replica catalog written → rc.yml")
print(f"      Registered: input_data.csv -> {WORK_DIR}/input_data.csv")
````

````python
# ────────────────────────────────────────────────────────────
# Cell 6: Workflow Definition — the DAG itself
# ────────────────────────────────────────────────────────────

# Create the workflow with automatic dependency inference.
# Pegasus traces which Files are inputs / outputs of each job
# and builds the DAG automatically — no manual dependency edges!
wf = Workflow("example-pipeline", infer_dependencies=True)

# ── Declare all intermediate Files ──────────────────────────
# These File objects are the glue that links jobs together.
# When Job A produces a File and Job B consumes it, Pegasus
# knows B depends on A.

# Preprocessing outputs
clean_data  = File("clean_data.csv")
data_stats  = File("data_stats.json")

# Model outputs (3 parallel models)
model_lr    = File("model_linear.pkl")
metrics_lr  = File("metrics_linear.json")

model_rf    = File("model_rf.pkl")
metrics_rf  = File("metrics_rf.json")

model_nn    = File("model_nn.pkl")
metrics_nn  = File("metrics_nn.json")

# Final outputs (stage_out=True → transferred to output site)
final_report = File("final_report.txt")
best_model   = File("best_model.pkl")

print("All File objects created —", len(wf.jobs), "jobs so far")
print("Files:")
for f in [clean_data, data_stats, model_lr, metrics_lr, model_rf, metrics_rf, model_nn, metrics_nn, final_report, best_model]:
    print(f"  • {f.lfn}")
````

````python
# ────────────────────────────────────────────────────────────
# Cell 7: Job 1 — Preprocessing
# ────────────────────────────────────────────────────────────

job_preprocess = Job(preprocess_trans)
job_preprocess.add_inputs(input_file)
job_preprocess.add_outputs(clean_data, data_stats)
job_preprocess.add_args(
    "--input", input_file,
    "--output_data", clean_data,
    "--output_stats", data_stats
)
wf.add_jobs(job_preprocess)

print(f"[Job] Preprocess added")
print(f"      Inputs:  {[f.lfn for f in job_preprocess.inputs]}")
print(f"      Outputs: {[f.lfn for f in job_preprocess.outputs]}")
````

````python
# ────────────────────────────────────────────────────────────
# Cell 8: Jobs 2a, 2b, 2c — Parallel Model Training
# ────────────────────────────────────────────────────────────
# These three jobs share the same transformation ("train")
# but take different --model_type arguments. They all depend
# on clean_data from the preprocess step.
# Pegasus automatically detects the dependency through the File.

# Job 2a: Linear Regression
job_lr = Job(train_trans)
job_lr.add_inputs(clean_data)
job_lr.add_outputs(model_lr, metrics_lr)
job_lr.add_args(
    "--input", clean_data,
    "--model_type", "linear_regression",
    "--output_model", model_lr,
    "--output_metrics", metrics_lr
)
wf.add_jobs(job_lr)

# Job 2b: Random Forest
job_rf = Job(train_trans)
job_rf.add_inputs(clean_data)
job_rf.add_outputs(model_rf, metrics_rf)
job_rf.add_args(
    "--input", clean_data,
    "--model_type", "random_forest",
    "--output_model", model_rf,
    "--output_metrics", metrics_rf
)
wf.add_jobs(job_rf)

# Job 2c: Neural Network
job_nn = Job(train_trans)
job_nn.add_inputs(clean_data)
job_nn.add_outputs(model_nn, metrics_nn)
job_nn.add_args(
    "--input", clean_data,
    "--model_type", "neural_network",
    "--output_model", model_nn,
    "--output_metrics", metrics_nn
)
wf.add_jobs(job_nn)

print("[Jobs] 3 parallel training jobs added")
print(f"       Linear Regression : {model_lr.lfn}, {metrics_lr.lfn}")
print(f"       Random Forest     : {model_rf.lfn}, {metrics_rf.lfn}")
print(f"       Neural Network    : {model_nn.lfn}, {metrics_nn.lfn}")
````

````python
# ────────────────────────────────────────────────────────────
# Cell 9: Job 3 — Evaluation (Fan-in Merge)
# ────────────────────────────────────────────────────────────
# This job collects ALL model outputs and produces the final
# report. Because it consumes files from all three training
# jobs, Pegasus knows it must wait for all three to finish.
# 
# stage_out=True marks these as final deliverables — they
# will be transferred to the output site (e.g., local storage).

job_evaluate = Job(evaluate_trans)
job_evaluate.add_inputs(
    clean_data,
    model_lr, metrics_lr,
    model_rf, metrics_rf,
    model_nn, metrics_nn,
)
job_evaluate.add_outputs(
    final_report,
    best_model,
    stage_out=True   # ← Final outputs staged to output site
)
job_evaluate.add_args(
    "--input_data", clean_data,
    "--model_lr", model_lr, "--metrics_lr", metrics_lr,
    "--model_rf", model_rf, "--metrics_rf", metrics_rf,
    "--model_nn", model_nn, "--metrics_nn", metrics_nn,
    "--output_report", final_report,
    "--output_best", best_model
)
wf.add_jobs(job_evaluate)

print("[Job] Evaluation (fan-in merge) added")
print(f"      Inputs:  {[f.lfn for f in job_evaluate.inputs]}")
print(f"      Outputs: {[f.lfn for f in job_evaluate.outputs]}  (stage_out=True)")
````

````python
# ────────────────────────────────────────────────────────────
# Cell 10: Register catalogs and write workflow YAML
# ────────────────────────────────────────────────────────────

# Attach all catalogs to the workflow
wf.add_site_catalog(sc)
wf.add_transformation_catalog(tc)
wf.add_replica_catalog(rc)

# Write the workflow definition (workflow.yml) plus catalogs
wf.write()

print("=" * 60)
print("  WORKFLOW GENERATED SUCCESSFULLY")
print("=" * 60)
print(f"  Location: {WORK_DIR}")
print(f"  Workflow: {wf.name}")
print(f"  Total jobs: {len(wf.jobs)}")
print()
print("  DAG structure:")
print(f"    1. Preprocess  →  {clean_data.lfn}, {data_stats.lfn}")
print(f"    2. Train x3     →  (3 parallel models)")
print(f"    3. Evaluate     →  (fan-in merge)")
print()
print("  Generated files:")
print(f"    • workflow.yml")
print(f"    • sites.yml")
print(f"    • tc.yml")
print(f"    • rc.yml")
````

````python
# ────────────────────────────────────────────────────────────
# Cell 11: Inspect the DAG — verify dependency inference
# ────────────────────────────────────────────────────────────

print("=" * 60)
print("  DAG DEPENDENCY ANALYSIS")
print("=" * 60)

for job in wf.jobs:
    parents = []
    for other in wf.jobs:
        if other is job:
            continue
        # If any output of 'other' is an input of 'job', it's a parent
        other_output_lfns = {f.lfn for f in other.outputs}
        job_input_lfns    = {f.lfn for f in job.inputs}
        if other_output_lfns & job_input_lfns:
            parents.append(other.name)
    
    print(f"\n  Job: {job.name}")
    print(f"    Transformation: {job.transformation}")
    print(f"    Inputs:  {[f.lfn for f in job.inputs]}")
    print(f"    Outputs: {[f.lfn for f in job.outputs]}")
    if parents:
        print(f"    Depends on: {parents}")
    else:
        print(f"    Depends on: (none — root job)")

print()
print("✓ Pegasus automatically inferred all dependencies via File sharing!")
````

````python
# ────────────────────────────────────────────────────────────
# Cell 12: Plan the workflow (pegasus-plan)
# ────────────────────────────────────────────────────────────
# This converts the abstract workflow into an executable plan.
# Uncomment the lines below to actually run pegasus-plan.
#
# To plan only (no submission):
#   !pegasus-plan --sites local --output-sites local workflow.yml
#
# To plan AND submit immediately:
#   !pegasus-plan --sites local --output-sites local --submit workflow.yml

print("=" * 60)
print("  PLANNING COMMAND")
print("=" * 60)
print()
print("To plan the workflow, run:")
print(f"  pegasus-plan --sites local --output-sites local {WORK_DIR}/workflow.yml")
print()
print("To plan and submit:")
print(f"  pegasus-plan --sites local --output-sites local --submit {WORK_DIR}/workflow.yml")
print()
print("To monitor:")
print("  pegasus-status <workflow_dir>")
print("  pegasus-analyzer <workflow_dir>")
````
